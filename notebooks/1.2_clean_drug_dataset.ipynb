## Data Cleaning

```python
# Drop the last two columns
drug_data = drug_data.iloc[:, :-2]
```

The last two columns `Chg_Avg_Spnd_Per_Dsg_Unt_21_22` and `CAGR_Avg_Spnd_Per_Dsg_Unt_18_22` were dropped as they are not needed for further reshaping of the dataset.

```python
# Define the columns that are repeated for each year
metrics = [
    "Tot_Spndng",
    "Tot_Dsg_Unts",
    "Tot_Clms",
    "Tot_Benes",
    "Avg_Spnd_Per_Dsg_Unt_Wghtd",
    "Avg_Spnd_Per_Clm",
    "Avg_Spnd_Per_Bene",
    "Outlier_Flag",
]

# Melt the dataset
long_format = drug_data.melt(
    id_vars=["Brnd_Name", "Gnrc_Name", "Tot_Mftr"],  # Keep these as identifiers
    value_vars=[f"{metric}_{year}" for year in range(2018, 2023) for metric in metrics],
    var_name="Metric_Year",
    value_name="Value",
)

# Extract Year and Metric separately
long_format["Year"] = long_format["Metric_Year"].str.extract(r"(\d{4})")
long_format["Metric"] = long_format["Metric_Year"].str.extract(r"(.+)_\d{4}")

# Drop the original Metric_Year column
long_format = long_format.drop(columns=["Metric_Year"])

# Pivot the table so each metric becomes a column
drug_data = long_format.pivot_table(
    index=["Brnd_Name", "Gnrc_Name", "Tot_Mftr", "Year"],
    columns="Metric",
    values="Value",
).reset_index()

# Print transformed data
drug_data.head()
```

The dataset has been reshaped and now includes the following:

- **Identifiers**: `Brnd_Name`, `Gnrc_Name`, `Tot_Mftr`, and `Year`
- **Metrics**: `Tot_Spndng`, `Tot_Dsg_Unts`, `Tot_Clms`, `Tot_Benes`, `Avg_Spnd_Per_Dsg_Unt_Wghtd`, `Avg_Spnd_Per_Clm`, `Avg_Spnd_Per_Bene`, and `Outlier_Flag`

Each metric is now represented as a separate column for each year, making it easier to analyze and compare data across different years.

### Rename columns
